```python
def get_token():
    auth_base64 = base64.b64encode(client_id.encode() + b':' + client_secret.encode()).decode("utf-8")

    url = "https://accounts.spotify.com/api/token"
    headers = {
        "Authorization": "Basic " + auth_base64,
        "Content-type": "application/x-www-form-urlencoded",
    }
 

    data = {
            "grant_type": "authorization_code", 
            "code": code,
            "redirect_uri": "http://localhost:3000",
            }
    result = post(url, headers=headers, data=data)
    json_result  = json.loads(result.content) #load script
  
    token = json_result["access_token"]
    return token
```

```python
def get_auth_header(token):
    return {
    "Authorization": "Bearer " + token,
    "Content-Type": "application/json"
}


def shorter_url(url = ""):
    full_url = BASIC_URL + url
    obj = get(full_url, headers = get_auth_header(token))
    print(obj)
    response = obj.json()
    return response

def get_my_top_items(type = "tracks", limit = 50, offset = 0, term = "medium_term") -> dict:
    obj = shorter_url(url = f"me/top/{type}?time_range={term}&limit={limit}&offset={offset}")
    return obj


def get_my_top_album_stats(type = "albums", limit = 50, offset = 0, term = "medium_term") -> dict:
    obj = shorter_url(url = f"me/{type}?time_range={term}&limit={limit}&offset={offset}")
    return obj


def get_track(id):
    resp = shorter_url(url = f"tracks/{id}")
    print(resp.json())

def get_me_stats():
    obj = shorter_url(url = f"me")
    return obj


def get_top_audio_features(id: str) -> dict:
    response = shorter_url(url= "audio-features?{id}")
    return response

```
